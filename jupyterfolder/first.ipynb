```python
import matplotlib.pyplot as plt
import numpy as np
import mplcursors  # For hover information

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
line, = plt.plot(x, y, label='sin(x)', color='blue')

# Add title and labels
plt.title('Interactive Plot with Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()

# Enable interactive cursor (hover to see data points)
mplcursors.cursor(line, hover=True)

# Show the plot
plt.show()
```
